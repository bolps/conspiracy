```python
df_full.columns
```

```python
import numpy as np
```

```python
tipi_cols = list(df_full.filter(like='TIPI',axis=1).columns)

for column in tipi_cols:
    df_full[column] = df_full[column].replace(0, np.nan)
    
df_full = df_full.dropna(subset=tipi_cols)
```

```python
# computing scores for TIPI (personality) scale and subscales according to the codebook. Note: Some of the items are reversed (formula for reversing: reverse = (number_of_levels + 1) - raw_value).
reverse = lambda x : (8-x)
df_full['TIPI_Extraversion'] = (df_full['TIPI1'] + reverse(df_full['TIPI6']))/2
df_full['TIPI_Agreeableness'] = (reverse(df_full['TIPI2']) + df_full['TIPI7'])/2
df_full['TIPI_Conscientiousness'] = (df_full['TIPI3'] + reverse(df_full['TIPI8']))/2
df_full['TIPI_Emotional_Stability'] = (reverse(df_full['TIPI4']) + df_full['TIPI9'])/2
df_full['TIPI_Openness'] = (df_full['TIPI5'] + reverse(df_full['TIPI10']))/2
```

```python
df_full['SURV_CheckStraightliners_TIPI'] = df_full[['TIPI1','TIPI2','TIPI3','TIPI4','TIPI5','TIPI6','TIPI7','TIPI8','TIPI9','TIPI10']].std(axis=1)
df_full['SURV_CheckStraightliners_TIPI'] = df_full['SURV_CheckStraightliners_TIPI'] == 0
```

```python
df_full['SURV_ValidityCheck_Errors'] = df_full['VCL6'] + df_full['VCL9'] + df_full['VCL12']
df_full = df_full.drop(['VCL1','VCL2','VCL3','VCL4','VCL5','VCL6','VCL7','VCL8','VCL9','VCL10','VCL11','VCL12','VCL13','VCL14','VCL15','VCL16'], axis=1)
```

```python
validity_mask = df_full['SURV_ValidityCheck_Errors'] == 0
```

```python
straightliners_mask = df_full['SURV_CheckStraightliners_TIPI'] == False
```

```python
df_clean = df_full[validity_mask & straightliners_mask]
```

```python
df_clean = df_clean.drop(['SURV_CheckStraightliners_TIPI', 'SURV_ValidityCheck_Errors'],axis=1)
```

```python
df_clean.columns
```

```python
df_clean
```

```python
# renaming columns for better organization.
# I logically divided data in blocks: RAW (raw GCBS and TIPI data), DEMO (demographics), GCBS (gcbs scores), TIPI (tipi scores) and SURV (survey related info)
df_clean.columns = [('RAW_'+column) for column in df_clean.columns[:df_clean.columns.get_loc('TIPI10')+1]] + [('DEMO_'+column) for column in df_clean.columns[df_clean.columns.get_loc('TIPI10')+1:df_clean.columns.get_loc('major')+1]] + list(df_clean.columns[df_clean.columns.get_loc('major')+1:])
```

```python
df_clean
```

```python
df_clean.columns = ['SOU_dataset' if x=='dataset' else x for x in df_clean.columns]
```

```python
df_clean
```

```python
df_clean.to_csv('TIPI_openpsychometrics_collection.csv', index=False)
```
